```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset_Facebook.csv", sep=';')
```

```python
# Display settings for proper structured output
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 3000)
```

```python
# 1. Display Dataset
display(df.head())
```

```python
# =========================================================
# a. Create Data Subsets
# =========================================================
```

```python
# Select Specific Columns
subset_cols = df[['Type', 'like']]
```

```python
# Display Selected Columns
display(subset_cols)
```

```python
# Select Rows Using Index
subset_rows = df.iloc[10:30]
```

```python
# Display Selected Rows
display(subset_rows)
```

```python
# Conditional Subset
subset_cond = df[df['like'] > 100]
```

```python
# Display Conditional Subset
display(subset_cond)
```

```python
# =========================================================
# b. Merge Data
# =========================================================
```

```python
# Create Sample Subsets
df1 = df.iloc[0:20]
df2 = df.iloc[20:40]
```

```python
# Merge Data
merged_df = pd.concat([df1, df2])
```

```python
# Display Merged Data
display(merged_df)
```

```python
# =========================================================
# c. Sort Data
# =========================================================
```

```python
# Sort Data by Likes in Descending Order
sorted_df = df.sort_values(by='like', ascending=False)
```

```python
# Display Sorted Data
display(sorted_df.head(20))
```

```python
# =========================================================
# d. Transposing Data
# =========================================================
```

```python
# Transpose Dataset
transposed_df = df.head().T
```

```python
# Display Transposed Data
display(transposed_df)
```

```python
# =========================================================
# e. Shape and Reshape Data
# =========================================================
```

```python
# Display Shape
print("Shape of Dataset:", df.shape)


# Convert Numeric Data into NumPy Array
arr = df.select_dtypes(include=np.number).to_numpy()

# Display Array
print(arr)
```

```python
# Reshape Array into Single Dimension
reshaped_arr = arr.reshape(-1)

# Display Reshaped Array
print(reshaped_arr)
```

```python
# Reshape Array into 2 Columns
reshaped_arr = arr.reshape(-1, 2)

# Display Reshaped Array
print(reshaped_arr)

1. # -1 Python automatically calculates number of rows.

2. # 2 Creates exactly 2 columns.
```
