```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
air_df = pd.read_csv(
    "AirQuality_visualization.csv",
    sep=';'
)

heart_df = pd.read_csv("heart.csv")
```

```python
print("Air Quality Dataset")

display(air_df.head())
```

```python
print("Heart Dataset")

display(heart_df.head())
```

```python
print(air_df.columns)

print(heart_df.columns)
```

```python
#line plot
# Put numeric column here

col = 'T'

plt.figure(figsize=(10,8))

plt.plot(
    air_df[col][:50]
)

plt.title("Temperature Variation")

plt.xlabel("Index")

plt.ylabel(col)

plt.show()
```

```python
print(air_df.columns)
```

```python
#HISTOGRAM
plt.figure(figsize=(8,5))

plt.hist(
    air_df[col].dropna(),
    bins=20
)

plt.title("Humidity Distribution")

plt.xlabel(col)

plt.ylabel("Frequency")

plt.show()

# plt.hist() -> histogram

# bins -> number of intervals

# dropna() -> removes null values

# histogram -> frequency distribution
```

```python
#SCATTER PLOT
num_cols = air_df.select_dtypes(include='number').columns

x = num_cols[0]
y = num_cols[1]

plt.figure(figsize=(10,5))

plt.scatter(
    air_df[x],
    air_df[y]
)

plt.title("Scatter Plot")

plt.xlabel(x)

plt.ylabel(y)

plt.show()
```

```python
#box plot
# Put your numeric column name here

# Put your numeric column here

col = 'age'

plt.figure(figsize=(10,5))

sns.boxplot(
    data=heart_df[col]
)

plt.title(f"{col} Boxplot")

plt.xlabel(col)

plt.grid()

plt.show()
```

```python
#count plot
col = 'sex'

plt.figure(figsize=(8,5))

sns.countplot(
    x=heart_df[col]
)

plt.title(f"{col} Count Plot")

plt.xlabel(col)

plt.ylabel("Count")

plt.show()
```

```python
# Heatmap

plt.figure(figsize=(10,8))

sns.heatmap(
    heart_df.corr(numeric_only=True),
    annot=True
)

plt.title("Heatmap")

plt.show()
```

```python
# Pairplot

sns.pairplot(
    heart_df.select_dtypes(include='number').iloc[:,0:4]
)

plt.show()
```
