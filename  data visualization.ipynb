```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import squarify
```

```python
# Load Iris Dataset
iris = sns.load_dataset('iris')

# Display Dataset
display(iris.head())
```

```python
#a. 1D (Linear) Visualization

plt.plot(iris['sepal_length'])

plt.title('1D Visualization')

plt.xlabel('Index')
plt.ylabel('Sepal Length')

plt.show()
```

```python
#b. 2D (Planar) Visualization

sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=iris
)

plt.title('2D Visualization')

plt.show()
```

```python
#c. 3D (Volumetric) Visualization

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    iris['sepal_length'],
    iris['petal_length'],
    iris['petal_width']
)

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')

plt.title('3D Visualization')

plt.show()
```

```python
#d. Temporal Visualization

iris['sepal_length'].plot()

plt.title('Temporal Visualization')

plt.xlabel('Index')
plt.ylabel('Sepal Length')

plt.show()
```

```python
#e. Multidimensional Visualization

sns.pairplot(
    iris,
    hue='species'
)

plt.show()
```

```python
#f. Tree / Hierarchical Visualization

sizes = iris['species'].value_counts()

squarify.plot(
    sizes=sizes.values,
    label=sizes.index
)

plt.title('Tree Visualization')

plt.axis('off')

plt.show()
```

```python
#g. Network Visualization

G = nx.Graph()

G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('C', 'D')
G.add_edge('D', 'B')

nx.draw(
    G,
    with_labels=True,
    node_size=8000
)

plt.title('Network Visualization')

plt.show()
```

```python
# 1D -> one variable
# 2D -> x and y axis
# 3D -> three dimensions
# Temporal -> data over time
# Multidimensional -> multiple variables
# Tree -> hierarchical structure
# Network -> connected nodes
```
